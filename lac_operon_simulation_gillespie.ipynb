```python
#packeges
import roadrunner
import matplotlib.pylab as plt
```

```python
#we want to compare the same model changing the initial concentration of the lactose Z

#Here we simulate the model with Lactose=1000
rr1 = roadrunner.RoadRunner("https://www.ebi.ac.uk/biomodels/BIOMD0000000065") #we load the model from Biomodels
rr1.setIntegrator('gillespie')
rr1.model['Lactose'] = 1000
result1 = rr1.simulate(0, 1000,10200)
Z= result1[:,7]
time= result1[:,0]

```

```python
#Here we simulate the model with Lactose=0

rr2 = roadrunner.RoadRunner("https://www.ebi.ac.uk/biomodels/BIOMD0000000065")
rr2.setIntegrator('gillespie')
rr2.model['Lactose'] = 0
result2 = rr2.simulate(0,1000,10200)

Z_s= result2[:,7]
time_s= result2[:,0]

#print time_s
#print result2
```

```python
from matplotlib import rc, rcParams
rc('text', usetex=True)
rc('font',**{'family':'sans-serif','sans-serif':['Computer Modern Sans Serif']})        # Change all fonts to 'Computer Modern'
rc('xtick', labelsize=18)
rc('ytick', labelsize=18)
rcParams['legend.numpoints'] = 1
```

```python
#here we plot the simulations results (with Lactose=1000 and Lactose_s =0)


plt.figure(figsize=(10,7))
plt.plot(time, Z , label ='$Z$')
plt.plot(time_s, Z_s , label ='$Z_s$')
plt.legend(loc="upper right", fontsize='xx-large')
plt.xlabel("Time", fontsize=20)
plt.ylabel("Concentrations",  fontsize=20)
plt.savefig('stoch_2.pdf')
plt.show()
```
